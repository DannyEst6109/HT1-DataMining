```python
# Mineria de datos
# Hoja de trabajo 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv('movies.csv')
```

```python
def graphNumericDistribution(data, title, xlabel, ylabel):
  mean = round(data.mean(), 2)
  median = round(data.median(), 2)
  mode = round(data.mode(), 2)
  max_value = round(data.max(), 2)
  plt.figure()
  sns.histplot(data, kde=True)
  plt.title(title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)

  plt.axvline(mean, color='r', linestyle='--', label='Mean')
  plt.axvline(median, color='g', linestyle='--', label='Median')
  plt.ylim(0, data.max())
  plt.legend()

  plt.text(0.98, 0.82, 'Mean:'+str(mean), transform=plt.gca().transAxes, ha='right', va='top')
  plt.text(0.98, 0.76, 'Median:'+str(median), transform=plt.gca().transAxes, ha='right', va='top')
  plt.text(0.98, 0.70, 'Mode:'+str(mode[0]), transform=plt.gca().transAxes, ha='right', va='top')
  plt.text(0.98, 0.64, 'Max:'+str(max_value), transform=plt.gca().transAxes, ha='right', va='top')
```

```python
def frequencyTable(data, title, xlabel, ylabel):
  frequency = data.value_counts()
  df_frequency = frequency.reset_index() # Convert the Series to DataFrame
  df_frequency.columns = [xlabel, ylabel] # Rename the columns
  df_frequency.plot(kind='bar', x=xlabel, y=ylabel, title=title)
  plt.ylabel(ylabel)
  print(df_frequency)
```

```python
# 1. Movies budget
moviesBudget = movies['budget']/1000000
graphNumericDistribution(moviesBudget, 'Movies Budget Distribution', 'Movie', 'Budget (Million $)')
```
